```python
import datetime
import matplotlib.pyplot as plt 
import pylab
import yfinance as yf
import numpy as np 
import pandas as pd 
import sys
import warnings
if not sys.warnoptions:
    warnings.simplefilter('ignore')
plt.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv("C:/Users/Mikayla/Desktop/Test.csv", index_col = 0, parse_dates=True)
df = df['Close']
df = pd.DataFrame(df)
df.head()
```

```python
reg = np.polyfit(np.arange(len(df)),df['Close'], deg = 4)
Quad = np.polyval(reg, np.arange(len(df)))
df['Poly'] = Quad
df.head()
```

```python
df['Poly'].plot(figsize=(15,10))
pylab.plot(df['Close'], 'd', label= 'Data Points From Stock')
pylab.legend()
plt.show()
```
